# Training Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
x = dataset.drop('prognosis', axis=1) # Drop this column wise (axis=1)
y = dataset['prognosis']
```

```python
x
```

```python
y
```

```python
# make y into array
le = LabelEncoder()
le.fit(y)
Y = le.transform(y)
```

```python
Y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,Y, test_size=0.3, random_state=20)
# x will have 70% of the data and y will have 30% of the data for both training and testing
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

## Training Models

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
```

```python
# dictionary of models
models = {
    #support vector machine
    'SVC': SVC(kernel='linear'),
    #random forest specifies 100 trees in the forest increasing the trees can improve mode but make more complex and slow
    #random_state is a seed value to make the model reproducible
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42),
    #gradient boosting specifies 100 trees in the forest increasing the trees can improve mode but make more complex and slow
    #random_state is a seed value to make the model reproducible
    'GradientBoosting': GradientBoostingClassifier(n_estimators=100, random_state=42),
    #k-nearest neighbors specifies 5 neighbors to vote on the class of the data point larger k makes it smoother but less precise
    'KNN': KNeighborsClassifier(n_neighbors=5),
    #naive bayes is a simple model that assumes the features are independent and counts the frequency of each class
    'NaiveBayes': MultinomialNB()
}
```

```python
# Loop through the models dictionary items for training and testing

for model_name, model in models.items():
    # Training the model
    model.fit(x_train, y_train)

    # Testing the model
    predictions = model.predict(x_test)

    # Calculate the accuracy of the model
    accuracy = accuracy_score(y_test, predictions)

    # Calculate confusion matrix
    matrix = confusion_matrix(y_test, predictions)

    print(f"{model_name} accuracy: {accuracy}, confusion matrix: ")

    print(np.array2string(matrix, separator=', '))    
    print("\n" + "="*40 + "\n")
```
